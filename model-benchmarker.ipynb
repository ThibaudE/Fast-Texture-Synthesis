```python
import matplotlib.pyplot as plt
```

```python
import glob
import json

stats_files = glob.glob("checkpoint/*.json")

stats_data = []
for stat_file in stats_files:
    f = open(stat_file)
    stats_data = stats_data + json.load(f)
```

```python
stats_data[0]
```

```python
plt.title("Improved Progressive PO Training Loss")
plt.xlabel("Epoch")
plt.ylabel("Texture Loss")
plt.plot([epoch_data["loss"] for epoch_data in stats_data]+[4301.6, 4145, 3100.1, 3268.2, 2292.5, 2365.2])
```
